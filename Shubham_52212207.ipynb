Q. 2 Consider a dataset where some attributes are dependent, then find the relationships

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

dataset_url = 'https://raw.githubusercontent.com/MainakRepositor/Datasets/master/Loan%20Eligibility/loan-train.csv'
# df = pd.read_csv('supermarket_sales.csv')

df = pd.read_csv(dataset_url)

df = df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']]

# Impute missing values in 'LoanAmount' column
imputer = SimpleImputer(strategy='mean')
df['LoanAmount'] = imputer.fit_transform(df[['LoanAmount']])

# Regression Analysis
X = df[['ApplicantIncome', 'CoapplicantIncome']]
y = df['LoanAmount']
regressor = LinearRegression()
regressor.fit(X, y)

fig = plt.figure(figsize=(12, 6))

ax1 = fig.add_subplot(121)
ax1.scatter(df['ApplicantIncome'], df['LoanAmount'], color='blue')
ax1.set_title('ApplicantIncome vs. LoanAmount')
ax1.set_xlabel('ApplicantIncome')
ax1.set_ylabel('LoanAmount')

ax2 = fig.add_subplot(122)
ax2.scatter(df['CoapplicantIncome'], df['LoanAmount'], color='red')
ax2.set_title('CoapplicantIncome vs. LoanAmount')
ax2.set_xlabel('CoapplicantIncome')
ax2.set_ylabel('LoanAmount')

plt.tight_layout()
plt.show()

print("relationship is :", regressor.coef_[0], "* ApplicantIncome +", regressor.coef_[1], "* CoapplicantIncome  = LoanAmount" )

```

Q. 3 Given a set of vectors, find if the set is linearly independent or dependent

```python
import numpy as np

def is_linearly_independent(vectors):
    # Form the augmented matrix
    matrix = np.column_stack(vectors)
    
    # Apply Gaussian elimination
    _, pivots = np.linalg.qr(matrix)
    
    # Check if every column contains a pivot
    if len(pivots) == np.linalg.matrix_rank(matrix):
        print("The set of vectors is linearly independent.")
    else:
        print("The set of vectors is linearly dependent.") 


v1 = np.array([1, 2, 3])
v2 = np.array([2, 4, 7])
vectors = [v1, v2]
is_linearly_independent(vectors)
  


v1 = np.array([1, 2, 3])
v2 = np.array([2, 4, 6])
vectors_2 = [v1, v2]
is_linearly_independent(vectors_2)

    
```

Q. 4 What are the row and column spaces of a given set of vectors?

The row space and column space of a set of vectors that describe the span of the rows and columns of a matrix formed by those vectors, respectively.

**Row Space:** The row space of the set of vectors is the vector space spanned by the rows of the matrix formed by row vectors. In other words, it is the space spanned by all possible linear combinations of the rows of the matrix.

**Column Space:** The column space of the set of vectors is the vector space spanned by the columns of the matrix formed by column vectors. It is the space spanned by all possible linear combinations of the columns of the matrix.

The dimension of the vector space spanned by a set of vectors is determined by the number of linearly independent vectors in the set.

Q. 6 In case many solutions exist or none then, can we find a unique solution by adding some more properties to the solutions?  What could be such properties?

When a system of linear equations has infinitely many solutions or no solution at all, adding additional properties to the solutions can sometimes lead to a unique solution. These additional properties can be constraints or conditions imposed on the solutions that help narrow down the possibilities. Some additional properties are

**Additional Equation:** Adding another linear equation to the system that is independent of the existing equations can help uniquely determine the solution. This additional equation acts as a constraint on the solution space.

**Boundary Conditions:** In some cases, specifying boundary conditions can help determine a unique solution. These conditions can limit the possible solutions to a finite set.

**Optimization Criteria:** Instead of finding any solution, you can aim to find the solution that optimizes certain criteria, such as minimizing or maximizing a certain function subject to the given constraints.

**Regularization:** In the context of regression problems, regularization techniques such as Ridge or Lasso regression add penalty terms to the objective function to ensure a unique solution.

**Discretization:** Discretizing a continuous problem can sometimes lead to a unique solution. By dividing the problem domain into discrete intervals or grids, you can ensure a finite number of possible solutions.

Q. 9 Apply linear regression to solve some regression problems. (DO NOT USE DIRECT FUNCTION INSTEAD MAKE YOUR OWN)

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)


def linear_regression(X, y):
    n = len(X)
    
   
    mean_x = np.mean(X)
    mean_y = np.mean(y)
    
    
    numer = sum((X - mean_x) * (y - mean_y))
    denom = sum((X - mean_x) ** 2)
    
 
    m = numer / denom
    
   
    b = mean_y - m * mean_x
    
    return m, b


slope, intercept = linear_regression(X, y)


y_pred = slope * X + intercept


print("Slope (m):", slope)
print("Intercept (b):", intercept)


plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```

Q. 10 Check if you can identify all three cases for a given matrix equation Ax=b, and find all the solutions in case they exists
Unique Solution
Infinite Solution
No Solution

```python
import numpy as np

def gauss_elimination(A, b):
   
    A = A.astype(float)
    b = b.astype(float)
    
    
    augmented_matrix = np.hstack((A, b.reshape(-1, 1)))
    n = augmented_matrix.shape[0]
    
 
    for i in range(n):
        pivot_row = augmented_matrix[i]
        if pivot_row[i] == 0:  # Check if pivot element is zero
            continue
        for j in range(i + 1, n):
            factor = augmented_matrix[j][i] / pivot_row[i]
            augmented_matrix[j] -= factor * pivot_row
    
    # Back substitution
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        if augmented_matrix[i][i] == 0:  # Check if diagonal element is zero
            return "Infinite solutions"
        x[i] = (augmented_matrix[i][-1] - np.dot(augmented_matrix[i][i+1:-1], x[i+1:])) / augmented_matrix[i][i]
    
    return x

def solve_linear_system(A, b):
    # Check if the system is consistent
    if np.linalg.matrix_rank(A) != np.linalg.matrix_rank(np.hstack((A, b.reshape(-1, 1)))):
        print("No solution (Case 3)")
        return None
    
    
    if A.shape[0] == A.shape[1] and np.linalg.matrix_rank(A) == A.shape[0]:
        solution = gauss_elimination(A, b)
        print("Unique solution (Case 1)")
        return solution
    
   
    print("Infinite solutions (Case 2)")
    return gauss_elimination(A, b)

# Example 1: Unique solution
A_unique = np.array([[2, 1], 
                     [1, -1]])
b_unique = np.array([3, 1])
print("Example 1:")
solution_unique = solve_linear_system(A_unique, b_unique)
print("Solution:", solution_unique)

# Example 2: Infinite solutions
A_infinite = np.array([[1, 2, 3], 
                       [2, 4, 6], 
                       [3, 6, 9]])
b_infinite = np.array([6, 12, 18])
print("\nExample 2:")
solutions_infinite = solve_linear_system(A_infinite, b_infinite)
print("Solutions:", solutions_infinite)

# Example 3: No solution
A_none = np.array([[1, 2], 
                   [2, 4]])
b_none = np.array([3, 7])
print("\nExample 3:")
solution_none = solve_linear_system(A_none, b_none)
print("Solution:", solution_none)
```
