**Can we visualise a system of linear equations as linear combination of vectors?**

Ans: YES

**What are different visualisation on**

A. Equation view of simultaneous system of linear equation.

B. Vector view of simultaneous system of linear equations.

```python
# A
# Let us consider two equation
# 2x + 3y = 6-----1
# 4x - 2y = 2------2

import numpy as np
import matplotlib.pyplot as plt

def equation1(x):
  return (6-2*x)/3
def equation2(x):
  return (4*x - 2)/-2

x_values = np.linspace(-5, 5, 100)

plt.plot(x_values, equation1(x_values), label='2x + 3y = 6')
plt.plot(x_values, equation2(x_values), label='4x - 2y = 2')

intersection_x = 2
plt.plot(intersection_x, equation1(intersection_x), 'ro', label='Intersection Point')

plt.title('System of Linear Equations')
plt.xlabel('x')
plt.ylabel('y')

plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

plt.show()
```

```python
# B
# let us consider a system of linear equation in vector-matrix form:
# Ax = b
# A = coefficient matrix
# x = column vector of variable
# b = column vector on the right hand side

import numpy as np
import matplotlib.pyplot as plt


# Defining the coefficient matrix A and the constant vector b
A = np.array([[2, 3], [4, -2]])
b = np.array([6, 2])

x = np.linalg.solve(A, b) # solve for x using NumPy's linear algebra solver

plt.figure(figsize=(8, 6))
origin = np.zeros_like(b)
plt.quiver(*origin, *A[:, 0], color=['r'], scale=1, scale_units='xy', angles='xy')
plt.quiver(*origin, *A[:, 1], color=['b'], scale=1, scale_units='xy', angles='xy')
plt.quiver(*origin, *b, color=['g'], scale=1, scale_units='xy', angles='xy')

plt.scatter(x[0], x[1], color='purple', label='Solution')
plt.title('Vector View of Linear Equations')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()


```

**What is linear combination of vectors?**

Solution:
A linear combination of vectors is an expression constructed from a set of vectors by scaling each vector by a scalar and then summing the results.

Given vectors v1,v2,...,vn and scalars c1,c2,....,cn the linear combination is given by:
c1v1 + c2v2 + ... + cnvn

Each term civi represents the scalar ci multiplied by the corresponding vector vi. The linear combination combines these scaled vectors by addition.


**What does it mean by linearly independent vector?**

Solution:

A linear combination of vectors refers to the sum of scalar multiples of those vectors. Let's consider a set of vectors v1,v2, ... ,vn and corresponding scalars c1,c2, ...,cn. The linear combination of these vectors is given by:

**c1v1 + c2v2 + ... +cnvn**

c1,c2... are the scalar coeffients

v1,v2,... are vectors

A. Linearly dependent vectors:

A set of vectors is said to be linearly dependent if at least one vector in the set can be expressed as a linear combination of the others.

In other words, vectors v1,v2, ... ,vn are linearly dependent if there exist scalars c1,c2, ... ,cn not all zero such that

c1v1, c2v2, ... ,cnvn = 0, where 0 is the zero vector.

B. Linearly independent vectors:

A set of vectors is said to be linearly independent if the only way to satisfy c1.v1 + c2v2 + ... + cnvn = 0 is when all the scalar coefficient is 0.

means, c1 = c2 = ... = cn = 0 .



```python
# What is matrix multiplication in terms of linear combinations of columns?

import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

C = np.dot(A, B)

result_columns = [f"{A[:, i]} * {B[i, :]}" for i in range(A.shape[1])]

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nMatrix C (Result of A * B):")
print(C)
print("\nLinear combinations of columns:")
print(result_columns)

```

```python
# What is matrix multiplication in terms of linear combinations of rows?

import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

C = np.dot(A, B)

result_rows = [f"{A[i, :]} * {B[:, i]}" for i in range(A.shape[0])]

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nMatrix C (Result of A * B):")
print(C)

print("\nLinear combinations of rows:")
print(result_rows)
```

**Generate matrices that provide the right scaling factors so that row/column vectors add to a particular vector.**

Solution:

To generate matrices such that the row or column vectors scale to a particular vector, we can set up a system of linear equations and solve for the matrix elements.



```python
import numpy as np
from scipy.linalg import solve

w = np.array([5, 3])

A_unknowns = np.random.rand(2, 2)

equations = np.dot(A_unknowns, np.array([[1, 0], [0, 1]]))

scaling_factors = solve(equations, w) #used to find the scaling factors.

A = A_unknowns * scaling_factors[:, np.newaxis]

print("Target vector (w):", w)
print("Generated matrix (A):")
print(A)
```


**What kind of space is created by linear combinations of given vectors?**

Solution:
The kind of space created by the linear combinations of the given vectors depends on the specific vectors and their linear independence. If the vectors are linearly independent, the space spanned by their linear combinations forms a subspace of the vector space. If the vectors are linearly dependent, the space they span is less than the dimensionality of the vector space. The space created by linear combinations is known as the column space or image space of the matrix.


**Can we generate all the vectors of that dimension? What does this question mean?**

Solution:

In linear algebra, a vector space is a collection of vectors that satisfies certain properties, including closure under vector addition and scalar multiplication. The span of a set of vectors is the set of all possible linear combinations of those vectors.


When we ask whether we can generate all the vectors of a certain dimension, we are essentially asking whether the set of vectors spans the entire vector space of that dimension. In other words, can any vector in that space be expressed as a linear combination of the given vectors?


1. Yes, if the vectors are linearly independent:

If the given vectors are linearly independent and form a basis for the vector space, then any vector in that space can be expressed as a unique linear combination of those vectors. In this case, the span of the vectors is the entire vector space.

2. No, if the vectors are linearly dependent:

If the vectors are linearly dependent, it means that some vectors in the set can be expressed as linear combinations of others. In this case, the span of the vectors is a subspace of the entire vector space, and they cannot generate all vectors of that dimension.

```python
# How do we solve the system of linear equations methodologically?
# Gauss Elimination Method

import numpy as np

def gauss_elimination(A, b):
    augmented_matrix = np.column_stack((A.astype(float), b.astype(float)))  # Convert to float
    n = len(b)

    for i in range(n):
        factor = augmented_matrix[i, i]
        augmented_matrix[i, :] /= factor

        for j in range(i + 1, n):
            factor = augmented_matrix[j, i] / augmented_matrix[i, i]
            augmented_matrix[j, :] -= factor * augmented_matrix[i, :]

    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = augmented_matrix[i, -1] - np.dot(augmented_matrix[i, i+1:n], x[i+1:])

    return x

A = np.array([[2, 3, 1], [-4, 2, 3], [1, -1, 2]])
b = np.array([9, 1, 5])


solution = gauss_elimination(A, b)
print("Solution using Gauss Elimination:", solution)
```

```python
#Gauss-Jordan Elimination method

import numpy as np

def gauss_jordan_elimination(A, b):
    augmented_matrix = np.column_stack((A.astype(float), b.astype(float)))  # Convert to float
    m, n = augmented_matrix.shape

    for i in range(m):
        factor = augmented_matrix[i, i]
        augmented_matrix[i, :] /= factor

        for j in range(m):
            if i != j:
                factor = augmented_matrix[j, i]
                augmented_matrix[j, :] -= factor * augmented_matrix[i, :]

    solution = augmented_matrix[:, -1]

    return solution


A = np.array([[2, 3, 1], [-4, 2, 3], [1, -1, 2]])
b = np.array([9, 1, 5])

solution = gauss_jordan_elimination(A, b)
print("Solution using Gauss-Jordan Elimination:", solution)
```
